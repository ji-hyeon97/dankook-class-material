# Quiz : Pyplot basics

```python
import pandas as pd

# the data we'll use
data = pd.read_csv('https://raw.githubusercontent.com/DA4BAM/dataset/master/temperature.csv')
data.head(10)
```

# Q1. Draw the scatter plot

* x-axis : 'week' column in data
* y-axis : 'AvgTemp' column in data

```python
####################
## your code here ##
####################

import matplotlib.pyplot as plt
plt.figure(figsize=(7,7 ))
plt.scatter(data['week'], data['AvgTemp'])
plt.show()
```

# Q2. Draw the bar plot
* x-axis : 'year' column in data
* y_axis(height) : mean of 'AvgTemp' column in data by year


```python
####################
## your code here ##
####################
temp = data.groupby(by='year', as_index=False)['AvgTemp'].mean()

plt.figure(figsize=(7,5 ))
plt.bar(x=temp['year'], height=temp['AvgTemp'])
plt.show()
```
