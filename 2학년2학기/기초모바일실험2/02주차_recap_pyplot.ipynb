# Pyplot basics

**Pyplot for beginners**
1. Basic code structure
2. Histogram
3. Bar plot
4. Scatter plot
5. Line plot

```python
import matplotlib.pyplot as plt
import pandas as pd

url = 'https://raw.githubusercontent.com/RayleighKim/Example_datasets/master/Graduate_apply.csv'
df = pd.read_csv(url)
df.head()
```

# 1. Basic code structure

**Three steps**
1. Figure size
2. Plots that you want
3. Plt.show()

```python
# First line to set the figure size
plt.figure(figsize=(10,5)) # figsize=(width, height)

# Second line to plot
plt.hist( df['gre'] ) # histogram

# Third line to show plot
plt.show()

# This line seems unnecessary in google colaboratory,
# but it will be useful for other IDE
```

# 2. Histogram

```python
# First line to set the figure size
plt.figure(figsize=(10,5)) # figsize=(width, height)

# Second line to plot
plt.hist( df['gpa'], bins=20 )

# Third line to show plot
plt.show()
```

# 3. Bar plot

```python
# you need some experience with pandas
# df[['rank']].value_counts()
temp = df.groupby(by='rank', as_index=False)['gpa'].count()
temp.columns = ['rank', 'freq']
temp
```

```python
# First line to set the figure size
plt.figure(figsize=(10,5)) # figsize=(width, height)

# Second line to plot
plt.bar( x = temp['rank'], height = temp['freq'] )

# Third line to show plot
plt.show()
```

```python
# First line to set the figure size
plt.figure(figsize=(10,5)) # figsize=(width, height)

# Second line to plot
plt.bar( x = ['1','2', '3', '4'], height = temp['freq'] )

# Third line to show plot
plt.show()
```

```python
# First line to set the figure size
plt.figure(figsize=(10,5)) # figsize=(width, height)

# Second line to plot
plt.barh( y = ['1','2', '3', '4'], width = temp['freq'] )

# Third line to show plot
plt.show()
```

# 4. Scatter plot

```python
# First line to set the figure size
plt.figure(figsize=(8,8)) # figsize=(width, height)

# Second line to plot
plt.scatter(x=df['gre'], y=df['gpa'])

# Third line to show plot
plt.show()
```

# 5. Line Plot

```python
t = list(range(1,11))
y = [4,6,2,2,7,9,1,2,8, 10]

# First line to set the figure size
plt.figure(figsize=(8,4)) # figsize=(width, height)

# Second line to plot
plt.plot(t, y)

# Third line to show plot
plt.show()
```
