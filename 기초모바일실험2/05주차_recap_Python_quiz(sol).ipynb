# Q2. Write code that does the following:
* Input : 
```
str_sample = 'a b b b c c d'
```
* Output :
```
>>> { 'a' : 1, 'b' : 3, 'c' : 2, 'd' : 1 }
 ```
* You don't need to make a function

````python
####################
## your code here ##
####################

str_sample = 'a b b b c c d'

temp = str_sample.split()

dict_count = {}

for key in temp :
    dict_count[key] = dict_count.get(key, 0) + 1

print(dict_count)
````

# Q3. Make a list that contains the values of origin(the list below) if only the values is higher than equal to 6

````python
origin = [ 9 ,5,6,2,1,19,34,1,36,9,5,4,162,125,62,1,23,3,6,1,24,4]

####################
## your code here ##
####################

# answer I
temp = []
for value in origin :
    if value >=6:
        temp.append(value)
print(temp)

# answer II

temp2 = [ v for v in origin if v >=6  ]
print(temp2)
````
